```python
import pandas as pd
from datetime import timedelta
from datetime import date
import requests
from io import StringIO
```

```python
def mta_url_list(weeks):
    base_url = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt" #180630
    latest_date = date(2018, 6, 30)
    td = timedelta(days=7)
    return [str.format(base_url, (latest_date - td*i).strftime("%y%m%d")) for i in list(range(weeks))]
```

```python
urls = mta_url_list(10)
df = pd.concat(
    [pd.read_csv(StringIO(requests.get(url).content.decode('utf8'))) for url in urls],
        ignore_index=True)
df.describe()
```

```python
df['ENTRIES NEXT 4HRS'] = df['ENTRIES'].shift(-1) - df['ENTRIES']
df.tail(100)
```
